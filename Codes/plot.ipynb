```python
import torch
import math
import numpy as np
import matplotlib.pyplot as plt
import pickle
import random

plt.rcParams.update({'font.size': 13})
```

```python
entropy_avg_moe_s1 = torch.load('entropy_change_records/MoE_s1.pt')
entropy_avg_moe_linear_s1 = torch.load('entropy_change_records/MoE_linear_s1.pt')

ci_moe_s1 = torch.load('entropy_change_records/MoE_ci_s1.pt')
ci_moe_linear_s1 = torch.load('entropy_change_records/MoE_ci_linear_s1.pt')

entropy_avg_moe_s2 = torch.load('entropy_change_records/MoE_s2.pt')
entropy_avg_moe_linear_s2 = torch.load('entropy_change_records/MoE_linear_s2.pt')

ci_moe_s2 = torch.load('entropy_change_records/MoE_ci_s2.pt')
ci_moe_linear_s2 = torch.load('entropy_change_records/MoE_ci_linear_s2.pt')
```

```python
plt.rcParams.update({'font.size': 18, 'figure.figsize': (6,6), 'font.family':'Arial', 'axes.axisbelow':True})
x = np.arange(0,501,1)

plt.plot(x,entropy_avg_moe_linear_s1, label='Linear-1')
plt.fill_between(x, (entropy_avg_moe_linear_s1+ci_moe_linear_s1), (entropy_avg_moe_linear_s1-ci_moe_linear_s1),
                alpha=.1)

plt.plot(x,entropy_avg_moe_s1, label='Nonlinear-1')
plt.fill_between(x, (entropy_avg_moe_s1+ci_moe_s1), (entropy_avg_moe_s1-ci_moe_s1),
                alpha=.1)

plt.plot(x,entropy_avg_moe_linear_s2, label='Linear-2')
plt.fill_between(x, (entropy_avg_moe_linear_s2+ci_moe_linear_s2), (entropy_avg_moe_linear_s2-ci_moe_linear_s2),
                alpha=.1)

plt.plot(x,entropy_avg_moe_s2, label='Nonlinear-2')
plt.fill_between(x, (entropy_avg_moe_s2+ci_moe_s2), (entropy_avg_moe_s2-ci_moe_s2),
                alpha=.1)

plt.xlabel('Training Epochs')
plt.ylabel('Entropy')
plt.legend(loc=5) #

plt.savefig('entropy.pdf',bbox_inches='tight')
```
